```python
import tensorflow_datasets as tfds
from rich import print
import torch
```

```python
dataset, dataset_info = tfds.load(
                "coco_captions",
                split="val",
                shuffle_files=False,
                download=True,
                as_supervised=False,
                data_dir='.scratch/datasets/',
                with_info=True,
            )

dataset = list(dataset.as_numpy_iterator())

image = dataset[0]["image"]
objects = dataset[0]["objects"]
object_ids = objects["label"]
object_bboxes = objects["bbox"]
```

```python
from matplotlib import pyplot as plt
```

```python
plt.imshow(image)
print(object_bboxes)
```

```python
x_min = int(object_bboxes[0][0] * image.shape[0])
y_min = int(object_bboxes[0][1] * image.shape[1])
x_max = int(object_bboxes[0][2] * image.shape[0])
y_max = int(object_bboxes[0][3] * image.shape[1])
plt.imshow(image[
           x_min:x_max,
           y_min:y_max])
```
